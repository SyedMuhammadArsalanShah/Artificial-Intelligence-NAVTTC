```python
import numpy as np
def shuffle_and_split_data(data,test_ratio):
    shuffle_indices=np.random.permutation(len(data))
    # print(shuffle_indices)
    test_size=int(len(data)*test_ratio)
    test_ind= shuffle_indices[:test_size]
    train_ind=shuffle_indices[test_size:]
    return data.iloc[train_ind], data.iloc[test_ind]
    
```

```python
train,test= shuffle_and_split_data(df,0.4)
```

```python
train
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index, test_index in split.split(df, df["income_cat"]):
    strata_train= df.loc[train_index]
    strata_test= df.loc[test_index]

```

```python
import matplotlib.pyplot as plt
strata_test["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
```

```python
# Lets remove the income_cat column
for sett in (strata_train, strata_test):
    sett.drop("income_cat", axis=1, inplace=True)
    
```

```python
housing= strata_train.drop("median_house_value", axis=1)
housing_label= strata_train["median_house_value"].copy()


from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
housing_num=housing.select_dtypes(include=[np.number])
imputer.fit(housing_num)
x=imputer.transform(housing_num)
x
```
