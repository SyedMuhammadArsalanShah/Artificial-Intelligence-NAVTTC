```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier

data = pd.read_excel("iris_training.xlsx")
data.head(10)
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1]
y
```

```python
model= DecisionTreeClassifier()
x.values
```

```python
y.values
```

```python
model.fit(x.values,y.values)
```

```python
prediction=model.predict([[5,2,	3.5,1]])
prediction
```

```python
data_test = pd.read_excel("iris_testing.xlsx")
data_test.head(10)
```

```python
x_test=data_test.iloc[:,:-1]
x_test
```

```python
y_test=data_test.iloc[:,-1]
y_test
```

```python
prediction=model.predict(x_test.values)
prediction
```

```python
actual=y_test.values
actual
```

```python
count=0
for i in range (0,len(actual)):
    if actual[i]==prediction[i]:
        count=count+1
print((count*100)/len(actual))
```
