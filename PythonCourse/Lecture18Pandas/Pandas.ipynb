## Pandas Data Selection & Filtering



```python

# Selecting Columns
df["Name"]              # Single column (Series)
df[["Name", "Age"]]           # Multiple columns (DataFrame)

```

```python
# Selecting Rows
print(df.loc[0])                   # First row by label
df.iloc[0]                     # First row by position

```

```python
# Selecting Specific Rows & Columns
df.loc[0, "Name"]              # Row 0, column 'Name'
df.iloc[0, 1]                  # Row 0, column at index 1
df.loc[0:2, ["Name", "Age"]]   # Rows 0–2, selected columns
df.iloc[0:2, 0:2]              # Rows & cols by position

```

```python
# Fast Access
df.at[0, "Name"]               # Fast label-based access
df.iat[0, 1]                   # Fast position-based access

```

```python
# Filtering with Conditions
df[df["Age"] > 30]             # Simple condition
df[(df["Age"] > 25) & (df["City"] == "Karachi")]   # Multiple AND
df[(df["Name"] == "Nawaira") | (df["Age"] < 30)]     # Multiple OR

```

```python
# Query Method (SQL-like filtering)
df.query("Age > 25 and City == 'Karachi'")   # Basic query
col = "Age"
df.query(f"{col} > 25")                    # Dynamic column
age_limit = 30
df.query("Age > @age_limit")               # Using Python variable

# Special Query Rules
# - Use backticks for spaces in column names → df.query("`first name` == 'Usman'")
# - String values must be quoted → df.query("name == 'Laiba'")
# - Use and/or/not instead of &, |, ~
# - Case-sensitive → df.query("City == 'Karachi'")
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame with some missing + duplicate values
data = {
    "name": ["Nawaira", "Fatimah", "Laiba", "Laiba", None],
    "age": [25, 30, np.nan, 35, 25],
    "city": ["Karachi", "Islamabad", "Lahore", "Lahore", None],
    "gender": ["F", "F", "F", "F", "F"]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df, "\n")

# ---------------- Missing Values ----------------
print("Check Missing:\n", df.isnull().sum(), "\n")   # Count NaNs per column
print("Drop Rows with NaN:\n", df.dropna(), "\n")   # Drop rows with any NaN
print("Fill NaN with 0:\n", df.fillna(0), "\n")     # Replace NaN with 0
print("Fill 'age' NaN with mean:\n", df["age"].fillna(df["age"].mean()), "\n")

# ---------------- Duplicates ----------------
print("Duplicate Rows:\n", df.duplicated(), "\n")   # True where row is duplicate
print("Drop Duplicates:\n", df.drop_duplicates(), "\n")

# ---------------- String Operations ----------------
print("Lowercase Names:\n", df["name"].str.lower(), "\n")  
print("Cities Containing 'lahore':\n", df["city"].str.contains("lahore", case=False), "\n")

# ---------------- Type Conversions ----------------
df["age"] = df["age"].astype("Int64")   # Convert to integer (nullable Int64)
df["category"] = df["city"].astype("category")  # City as category type
print("Dtypes:\n", df.dtypes, "\n")

# ---------------- Applying Functions ----------------
df["age_group"] = df["age"].apply(lambda x: "Adult" if x >= 18 else "Minor")
gender_map = {"F": "Female", "M": "Male"}
df["gender"] = df["gender"].map(gender_map)       # Map M/F → Male/Female
df["city"] = df["city"].replace({"Karachi": "KHI", "Lahore": "LHR"})  # Replace city names

print("Final Cleaned DataFrame:\n", df)
```

```python

# Sample DataFrame
data = {
    "name": ["Nawaira", "Fatimah", "Laiba", "Ayesha"],
    "age": [25, 30, 25, 28],
    "city": ["Karachi", "Islamabad", "Lahore", "Multan"],
    "salary": [60000, 75000, 60000, 70000],
    "score": [88, 92, 88, 95]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df, "\n")

# -------- Sorting --------
print("Sort by Age (ascending):\n", df.sort_values("age"), "\n")
print("Sort by Age (descending):\n", df.sort_values("age", ascending=False), "\n")
print("Sort by Age & Salary:\n", df.sort_values(["age", "salary"]), "\n")

# Reset index after sorting
df_sorted = df.sort_values("age").reset_index(drop=True)
print("Reset Index:\n", df_sorted, "\n")

# Sort by Index
df_shuffled = df.sample(frac=1, random_state=1)   # shuffle rows randomly
print("Shuffled:\n", df_shuffled, "\n")
print("Sort by Index:\n", df_shuffled.sort_index(), "\n")

# -------- Ranking --------
df["rank_avg"] = df["score"].rank()                  # default average
df["rank_dense"] = df["score"].rank(method="dense")  # no gaps
print("Ranking with average & dense:\n", df, "\n")

# -------- Renaming --------
df.rename(columns={"name": "Name", "age": "Age"}, inplace=True)  # rename cols
df.rename(index={0: "row1", 1: "row2"}, inplace=True)            # rename index
print("Renamed DataFrame:\n", df, "\n")

# -------- Change Column Order --------
# Full reordering
df = df[["city", "Name", "Age", "salary", "score", "rank_avg", "rank_dense"]]
print("Reordered Columns:\n", df, "\n")

# Move specific column ("Name") to front
cols = ["Name"] + [col for col in df.columns if col != "Name"]
df = df[cols]
print("Move 'Name' to front:\n", df)
```

```python
import pandas as pd 

data = {
    "Name": ["Nawaira", "Usman", "Laiba", "Mufti Sahab"],
    "AI":   [10, 98, 99, 89],
    "ML":   [100, 98, 99, 89],
    "SQL":  [10, 98, 99, 89],   
}

df = pd.DataFrame(data)

# Melt: Wide to Long
df_melted = pd.melt(df, id_vars=["Name"], 
                    value_vars=["AI", "ML", "SQL"], 
                    var_name="Subject", value_name="Obtained Marks")

print(df_melted)
```

```python
# Pivot Table: Long to Wide
df_pivot = df_melted.pivot_table(index="Name", 
                                 columns="Subject", 
                                 values="Obtained Marks", 
                                 aggfunc="mean")  # mean by default

print(df_pivot)
```

```python


# ---- Data ----
data = {
    "Name":["Nawaira", "Usman", "Laiba", "Mufti Sahab"],
    "AI":[10, 98, 99, 89],
    "ML":[100, 98, 99, 89],
    "SQL":[10, 98, 99, 89],   
}
df = pd.DataFrame(data)

print(" Original DataFrame")
print(df, "\n")

# ---- Pivot Example ----
pivot = df.melt(id_vars="Name", value_vars=["AI","ML","SQL"], 
                var_name="Subject", value_name="Marks")
print(" Pivoted Data (Long Format)")
print(pivot, "\n")

# ---- GroupBy Example ----
group_mean = pivot.groupby("Subject")["Marks"].mean()
print(" GroupBy - Average Marks per Subject")
print(group_mean, "\n")

# ---- Aggregation Example ----
agg = pivot.groupby("Subject")["Marks"].agg(["mean", "max", "min", "count"])
print(" Aggregation (mean, max, min, count)")
print(agg, "\n")

# ---- Custom Aggregation Example ----
custom = pivot.groupby("Subject").agg(
    avg_marks=("Marks", "mean"),
    highest=("Marks", "max"),
    lowest=("Marks", "min")
)
print(" Custom Aggregation")
print(custom, "\n")

# ---- Transform Example ----
pivot["Subject Avg"] = pivot.groupby("Subject")["Marks"].transform("mean")
print(" Transform Example (Subject Average per row)")
print(pivot, "\n")

# ---- Filter Example ----
filtered = pivot.groupby("Subject").filter(lambda x: x["Marks"].mean() > 60)
print(" Filter Example (Subjects with avg > 60)")
print(filtered)
```

```python

# --- Departments DataFrame ---
data_dep = {
    "DeptId": [1, 2, 3, 4],
    "DeptName": ["HR", "Sales", "IT", "Finance"],
    "DeptHead": ["Usman", "KaleemUllah", "Nawaira", "Laiba"]
}
df_dep = pd.DataFrame(data_dep)
print(" Departments Table")
print(df_dep, "\n")

# --- Employees DataFrame ---
data_emp = {
    "ID": [1, 2, 3, 4],
    "Name": ["Hira", "Salman", "Inshaal", "Zaid"],
    "DeptId": [1, 2, 0, 0]   # 0 means no department assigned
}
df_emp = pd.DataFrame(data_emp)
print(" Employees Table")
print(df_emp, "\n")

# --- Merge Employees with Departments ---
print(" Inner Join (only matching DeptId)")
print(pd.merge(df_emp, df_dep, on="DeptId"), "\n")

print(" Left Join (all employees, missing depts = NaN)")
print(pd.merge(df_emp, df_dep, on="DeptId", how="left"), "\n")

print(" Right Join (all departments, even if no employee)")
print(pd.merge(df_emp, df_dep, on="DeptId", how="right"), "\n")

print(" Outer Join (all employees + all departments)")
print(pd.merge(df_emp, df_dep, on="DeptId", how="outer"))
```

```python


# --- Vertical Concatenation (rows) ---
df_emp1 = pd.DataFrame({"Name": ["Hira", "Salman"]})
df_emp2 = pd.DataFrame({"Name": ["Inshaal", "Zaid"]})

print("Vertical Concatenation (Employees):")
print(pd.concat([df_emp1, df_emp2]), "\n")

# --- Horizontal Concatenation (columns) ---
df_dep1 = pd.DataFrame({"DeptId": [1, 2]})
df_dep2 = pd.DataFrame({"DeptHead": ["Usman", "KaleemUllah"]})

print("Horizontal Concatenation (Departments):")
print(pd.concat([df_dep1, df_dep2], axis=1))
```

```python
import pandas as pd 

# --- Sample Data ---
data_emp = {
    "ID": [1, 2, 3, 4],
    "Name": ["Hira", "Salman", "Inshaal", "Zaid"],
    "DeptId": [1, 2, 3, 4]
}
df_emp = pd.DataFrame(data_emp)

data_dep = {
    "DeptId": [1, 2, 3, 4],
    "DeptName": ["HR", "Sales", "IT", "Finance"],
    "DeptHead": ["Usman", "KaleemUllah", "Nawaira", "Laiba"]
}
df_dep = pd.DataFrame(data_dep)

# --- CSV Files ---
df_emp.to_csv("employees.csv", index=False)   # Write to CSV
df_emp_from_csv = pd.read_csv("employees.csv")   # Read from CSV

# --- Excel Files ---
df_dep.to_excel("departments.xlsx", index=False)   # Write single sheet
df_dep_from_excel = pd.read_excel("departments.xlsx")  # Read Excel file

# Multiple sheets
with pd.ExcelWriter("company_report.xlsx") as writer:
    df_emp.to_excel(writer, sheet_name="Employees", index=False)
    df_dep.to_excel(writer, sheet_name="Departments", index=False)

# --- JSON Files ---
df_emp.to_json("employees.json", orient="records")  # Write JSON
df_emp_from_json = pd.read_json("employees.json")   # Read JSON

# --- Show Outputs ---
print("CSV Read:\n", df_emp_from_csv, "\n")
print("Excel Read:\n", df_dep_from_excel, "\n")
print("JSON Read:\n", df_emp_from_json)

# Note:
# Orient Options
# orient="records" → List of dictionaries (best for APIs, human readable).
# orient="split" → JSON with separate keys: index, columns, data.
# orient="index" → Dict of dicts, index values as keys
# orient="columns" → Dict of lists, column names as keys.
# orient="values" → Just raw nested lists (no labels).
```
